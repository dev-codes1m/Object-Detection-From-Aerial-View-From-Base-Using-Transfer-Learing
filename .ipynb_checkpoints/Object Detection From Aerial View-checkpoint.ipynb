## Generating the More Image using ImageData Generator

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                  shear_range=0.2,
                                  zoom_range=0.2,
                                  validation_split=0.1)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
training_set = train_datagen.flow_from_directory('D:/Dwarf AI/Train Data',target_size=(224,224),batch_size=32,class_mode='categorical',shuffle=True)
testset = test_datagen.flow_from_directory('D:/Dwarf AI/Test Data',
                                          target_size=(224,224),
                                          batch_size=32,
                                          class_mode='categorical',shuffle=True)
```

```python
folders = glob('D:/Dwarf AI/Train Data/*') # Detecting the Number of Folder in the sub Folder
```

```python
len(folders)
```

```python
vgg = VGG16(input_shape=(224,224,3),weights='imagenet',include_top = False)# Importing the Pretrained Model
```

```python
for layers in vgg.layers:
    layers.trainable = False # As We are Using the Pretrained Model hence the trainable Layer need not to be Trained.
```

```python
X = Flatten()(vgg.output)
X = Dense(100,activation='relu')(X) # Adding the Dense layer 
prediction = Dense(len(folders),activation='softmax')(X) # Using softmax Fucntion Becoz here we are Making Muliple Classification
```

```python
Model = Model(inputs=vgg.input,outputs = prediction)
```

```python
Model.summary()
```

```python
Model.compile(loss = 'categorical_crossentropy',optimizer = 'adam',metrics=['acc']) # Compliling The Model
```

```python
hist = Model.fit_generator(training_set,validation_data=testset,epochs=50,
                       validation_steps=len(testset),
                       steps_per_epoch =len(training_set),verbose=1) # training the Model
```

```python
from keras.preprocessing import image
img = image.load_img(r"D:/Dwarf AI/testimg.jpg",target_size=(224,224)) # Importing the Image To Test
```

```python
from  keras.preprocessing.image import img_to_array
x = image.img_to_array(img) # Converting the Image Into The Array so that it can be Accepted By the Model
```

```python
x = np.expand_dims(x,axis=0) # Getting the Maximum Pixel Of the Image To make clear Predictions
```

```python
x=x/255.0 # Normalization Of the image
```

```python
k = Model.predict(x)
```

```python
if k[0][0]==np.max(k):
    print('Car')
elif k[0][1]==np.max(k):
    print('MotorBike')
elif k[0][2]==np.max(k):
    print('Peoples')
elif k[0][3]==np.max(k):
    print('Trash')
```
